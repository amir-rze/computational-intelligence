```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
from sklearn import svm
```

Open CSV files(train & test)

```python
train_dataset = pd.read_csv('train.csv')
test_dataset = pd.read_csv('test.csv')
train_dataset
```

replace NaN values with the most frequent value.(according to https://analyticsindiamag.com/5-ways-handle-missing-values-machine-learning-datasets/)

```python
nan_value_columns = train_dataset.columns[train_dataset.isna().any()].tolist()
for column in nan_value_columns:
    train_dataset.loc[train_dataset[column].isnull(), column] = train_dataset[column].value_counts().idxmax()
nan_value_columns = test_dataset.columns[test_dataset.isna().any()].tolist()
for column in nan_value_columns:
    test_dataset.loc[test_dataset[column].isnull(), column] = test_dataset[column].value_counts().idxmax()

train_dataset
```

split lable from features

```python
train_y = train_dataset['Survived']
train_x = train_dataset.drop(columns=['Survived'])

test_x = test_dataset
train_y
```

```python
train_x
```

```python
train_x = train_x.apply(LabelEncoder().fit_transform)
test_x = test_x.apply(LabelEncoder().fit_transform)
train_x
```

```python
test_x
```

Standardization

```python
standard_train_x = preprocessing.StandardScaler().fit_transform(train_x)
standard_train_x
```

```python
standard_test_x = preprocessing.StandardScaler().fit_transform(test_x)
standard_test_x
```

SVM

```python
function = svm.SVC()
function.fit(standard_train_x, train_y)
result = function.predict(standard_test_x)
```

Result

```python
result
```

```python
test_dataset['Survived']=result
test_dataset
```
